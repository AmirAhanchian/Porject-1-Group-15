```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import hvplot.pandas
from scipy.stats import linregress
import scipy.stats
```

```python
country_data_df = pd.read_csv("../output_data/Cleaned_Countries_Data.csv", index_col="Country_ID")

country_data_df
```

```python
def linear_reg(x,y):

    (slope, intercept, rvalue, pvalue, stderr) = linregress(x['Real GDP Per Capita 2020'],y)
    regress_values = slope*x['Real GDP Per Capita 2020']+intercept
    line_eq = "y=" + str(round(slope,2))+"x +" + str(round(intercept,2))
    
    plt.scatter(x['Real GDP Per Capita 2020'], y)
    plt.plot(x['Real GDP Per Capita 2020'],regress_values,"r-")
    plt.annotate(line_eq,(0.2,0.22),xycoords='figure fraction', fontsize = 13, color = "black")

    r_squared = round(rvalue**2,2)
    plt.annotate(f'the r-value is {r_squared}',(0.2,0.18),xycoords='figure fraction', fontsize = 13, color = "black")
    plt.xlabel("Real GDP Per Capita 2020")
    print(f'the rsquare-value is {r_squared}')
```

---

## Compute Linear Regression for Each Relationship


Linear regression of GDP vs C02 Emissions:

```python
linear_reg(country_data_df,country_data_df['C02 Emissions (megatons)'])
plt.ylabel('C02 Emissions (megatons)')
plt.title(f'Real GDP Per Capita 2020 vs CO2 Emissions')


# make a map with lan long and size it with co2 emission
```

Linear regression of GDP vs Particulate Matter Emissions:

```python
linear_reg(country_data_df,country_data_df['Particulate Matter Emissions (micrograms per m3)'])
plt.ylabel('Particulate Matter Emissions (micrograms per cubic meter)')
plt.title(f'Real GDP Per Capita 2020 vs Particulate Matter Emissions (micrograms per m3)')
```

```python
linear_reg(country_data_df,country_data_df['% of Wasted Recycled Annually'])
plt.ylabel('Percent of Waste Recycled Annually (%)')
plt.title(f'Real GDP Per Capita 2020 vs Percent of Waste Recycled Annually')
```

```python
linear_reg(country_data_df,country_data_df['Urban Population (%)'])
plt.ylabel('Urban Population (%)')
plt.title(f'Real GDP Per Capita 2020 vs Urban Population (%)')
```
