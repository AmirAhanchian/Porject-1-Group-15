```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
```

```python
country_data_df = pd.read_csv("../output_data/Cleaned_Countries_Data.csv", index_col="Country_ID")

country_data_df
```

```python
def linear_reg(x,y):

    (slope, intercept, rvalue, pvalue, stderr) = linregress(x['Urban Population (%)'],y)
    regress_values = slope*x['Urban Population (%)']+intercept
    line_eq = "y=" + str(round(slope,2))+"x +" + str(round(intercept,2))
    
    plt.scatter(x['Urban Population (%)'], y)
    plt.plot(x['Urban Population (%)'],regress_values,"r-")
    plt.annotate(line_eq,(0.2,0.22),xycoords='figure fraction', fontsize = 13, color = "black")

    r_squared = round(rvalue**2,2)
    plt.annotate(f'the r-value is {r_squared}',(0.2,0.18),xycoords='figure fraction', fontsize = 13, color = "black")
    plt.xlabel("Urban Population (%)")
    print(f'the rsquare-value is {r_squared}')
```

---

## Compute Linear Regression for Each Relationship


Linear regression of GDP vs C02 Emissions:

```python
linear_reg(country_data_df,country_data_df['C02 Emissions (megatons)'])
plt.ylabel('C02 Emissions (megatons)')
plt.title(f'Urban Population (%) vs CO2 Emissions')
```

Linear regression of GDP vs Particulate Matter Emissions:

```python
linear_reg(country_data_df,country_data_df['Particulate Matter Emissions (micrograms per m3)'])
plt.ylabel('Particulate Matter Emissions (micrograms per cubic meter)')
plt.title(f'Urban Population (%) vs Particulate Matter Emissions (micrograms per m3)')
```

```python
linear_reg(country_data_df,country_data_df['% of Wasted Recycled Annually'])
plt.ylabel('Percent of Waste Recycled Annually (%)')
plt.title(f'Urban Population (%) vs Percent of Waste Recycled Annually')
```

```python
linear_reg(country_data_df,country_data_df['Real GDP Per Capita 2020'])
plt.ylabel('Real GDP Per Capita 2020')
plt.title(f'Urban Population (%) vs Real GDP Per Capita 2020')
```
