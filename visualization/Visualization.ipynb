```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
```

```python
country_data_df = pd.read_csv("../output_data/countries.csv", index_col="Country_ID")

country_data_df
```

```python
def linear_reg(x,y):

    (slope, intercept, rvalue, pvalue, stderr) = linregress(x['GDP (Adjusted to $US)'],y)
    regress_values = slope*x['GDP (Adjusted to $US)']+intercept
    line_eq = "y=" + str(round(slope,2))+"x +" + str(round(intercept,2))
    
    plt.scatter(x['GDP (Adjusted to $US)'], y)
    plt.plot(x['GDP (Adjusted to $US)'],regress_values,"r-")
    plt.annotate(line_eq,(0.2,0.22),xycoords='figure fraction', fontsize = 13, color = "black")

    r_squared = round(rvalue**2,2)
    plt.annotate(f'the r-value is {r_squared}',(0.2,0.18),xycoords='figure fraction', fontsize = 13, color = "black")
    plt.xlabel("GDP (Adjusted to $US)")
    print(f'the r-value is {r_squared}')
```

---

## Compute Linear Regression for Each Relationship


Linear regression of GDP vs C02 Emissions:

```python
linear_reg(country_data_df,country_data_df['C02 Emissions'])
plt.ylabel('CO2 Emissions (megatons)')
plt.title(f'Country total GDP vs CO2 Emissions')
```

Linear regression of GDP vs Particulate Matter Emissions:

```python
linear_reg(country_data_df,country_data_df['Particulate Matter Emissions'])
plt.ylabel('Particulate Matter Emissions (micrograms per cubic meter)')
plt.title(f'Country total GDP vs Particulate Matter Emissions')
```

```python
linear_reg(country_data_df,country_data_df['% of Waste Recycled Annually'])
plt.ylabel('Percent of Waste Recycled Annually (%)')
plt.title(f'Country total GDP vs Percent of Waste Recycled Annually')
```
