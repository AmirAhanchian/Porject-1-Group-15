```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read saved data
country_data_df = pd.read_csv("../output_data/countries.csv", index_col="Country_ID")

# Display sample data
country_data_df.head()
```

```python
coordsplit = [country_data_df['Geographical Coordinates'].str.split(', ', expand=True)]
coords = coordsplit[0]
lat = coords[0]
lon = coords[1]
```

```python
gdpsplit = [country_data_df['GDP (Adjusted to $US)'].str.split(' ', expand=True)]
gdps = gdpsplit[0]
gdp = gdps[0]
```

```python
wastegensplit = [country_data_df['Waste Generated Annually'].str.split(' ', expand=True)]
wgs = wastegensplit[0]
wastegen = wgs[0]

wasterecsplit = [country_data_df['Waste Recycled Annually'].str.split(' ', expand=True)]
wrs = wasterecsplit[0]
wasterec = wrs[0]

perrecsplit = [country_data_df['% of Waste Recycled Annually'].str.split(' ', expand=True)]
prs = perrecsplit[0]
perrec = prs[0]
```

```python
pmemssplit = [country_data_df['Particulate Matter Emissions'].str.split(' ', expand=True)]
pes = pmemssplit[0]
pmems = pes[0]

co2emssplit = [country_data_df['C02 Emissions'].str.split(' ', expand=True)]
co2s = co2emssplit[0]
co2ems = co2s[0]

memssplit = [country_data_df['Methane  Emissions'].str.split(' ', expand=True)]
mes = memssplit[0]
mems = mes[0]
```

```python
envagtot = []
y = 0

while y < country_data_df['Country'].count():
    envagnums = len(country_data_df['Environmental Agreements (party to)'].str.split(', ')[y])
    envagtot.append(envagnums)  
    y += 1

print(envagtot)
```

```python
envagnrtot = []
x = 0

while x < country_data_df['Country'].count():
    envagnrnums = len(country_data_df['Environmental Agreements (signed but not ratified)'].str.split(', ')[x])
    envagnrtot.append(envagnrnums)  
    x += 1

print(envagnrtot)
```

```python
ebffsplit = [country_data_df['Electricity by Fossil Fuels'].str.split(' ', expand=True)]
ebffs = ebffsplit[0]
ebff = ebffs[0]

ebnsplit = [country_data_df['Electricity by Nuclear'].str.split(' ', expand=True)]
ebns = ebnsplit[0]
ebn = ebns[0]

ebssplit = [country_data_df['Electricity by Solar'].str.split(' ', expand=True)]
ebss = ebssplit[0]
ebs = ebss[0]

ebwsplit = [country_data_df['Electricity by Wind'].str.split(' ', expand=True)]
ebws = ebwsplit[0]
ebw = ebws[0]

ebhesplit = [country_data_df['Electricity by Hydroelectricty'].str.split(' ', expand=True)]
ebhes = ebhesplit[0]
ebhe = ebhes[0]

ebtwsplit = [country_data_df['Electricity by Tide and Wave'].str.split(' ', expand=True)]
ebtws = ebtwsplit[0]
ebtw = ebtws[0]

ebgtsplit = [country_data_df['Electricity by Geothermal'].str.split(' ', expand=True)]
ebgts = ebgtsplit[0]
ebgt = ebgts[0]

ebbwsplit = [country_data_df['Electricity by Geothermal'].str.split(' ', expand=True)]
ebbws = ebbwsplit[0]
ebbw = ebbws[0]
```

```python
upsplit = [country_data_df['Urban Population (%)'].str.split(' ', expand=True)]
ups = upsplit[0]
up = ups[0]

nmsplit = [country_data_df['Net Migration'].str.split(' ', expand=True)]
nms = nmsplit[0]
nm = nms[0]
```

```python
cleaned_df = pd.DataFrame({'Country': country_data_df['Country'], 
                        'Lat': lat, 
                        'Lon': lon, 
                        'GDP': gdp,
                        'Annual Waste Generated (tons)': wastegen,
                        'Annual Waste Recycled (tons)': wasterec,
                        'Waste Recycled %': perrec,
                        'Particulate Matter Emissions (micrograms per cubic meter)': pmems,
                        'C02 Emissions (megatons)': co2ems,
                        'Methane  Emissions (megatons)': mems,
                        "Environmental Agreements (party to)": envagtot,
                        "Environmental Agreements (signed but not ratified)": envagnrtot,
                        'Electricity by Fossil Fuels': ebff,
                        "Electricity by Nuclear": ebn,
                        "Electricity by Solar": ebs,
                        "Electricity by Wind": ebw,
                        "Electricity by Hydroelectricty": ebhe,
                        "Electricity by Tide and Wave": ebtw,
                        "Electricity by Geothermal": ebgt,
                        "Electricity by Biomass and Waste": ebbw, 
                        "Urban Population (%)": up,
                        "Net Migration (/1000 Population)": nm
                       })
cleaned_df
```

```python
# Testing if a graph works off the data

# Create plot
plt.scatter(cleaned_df['GDP'],cleaned_df['Waste Recycled %'],c="tab:blue")

# Incorporate the other graph properties
plt.title(f'Country GDP  vs. Waste Recycled')
plt.xlabel("GDP")
plt.ylabel("Waste Recycled (%)")
plt.grid(False)

# Save the figure
plt.savefig("../output_data/Fig1.png")

# Show plot
plt.show()
```
